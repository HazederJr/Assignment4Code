```python
import matplotlib.pyplot as plt
import numpy as np

#Our estimate. We predict it tends towards 2/3, but we can edit this to a function later
def f_theoretical(n):
    return (2/3)

#We compute our theoretical probabilities for steps 1 to 30
steps = np.arange(1, 31)
theoretical_probs = [f_theoretical(n) for n in steps]

#Simulate empirical probabilities for each step
def simulate_probability(n, trials=10000):
    hits = 0
    for _ in range(trials):
        pos = 1
        # Continue until we reach or pass step n.
        while pos < n:
            pos += np.random.choice([1, 2])
        if pos == n:
            hits += 1
    return hits / trials

empirical_probs = [simulate_probability(n) for n in steps]

plt.figure(figsize=(8, 6))
plt.plot(steps, theoretical_probs, label='$\mathbb{P}(n) = 2/3$',  linestyle='-')
plt.scatter(steps, empirical_probs, color='red', label='Empirical estimate', zorder=5)
plt.xlabel("Step Number")
plt.ylabel("Probability of landing exactly on step n")
plt.title("Probability of Landing Exactly on Step n")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Our estimate. We predict it tends towards 2/3, but we can edit this to a function later
def f_theoretical(n):
    theta = np.arccos(-np.sqrt(3)/3)
    return 0.5 + 0.5 * (1/np.sqrt(3))**(n-1) * np.cos((n-1)*theta)

#We compute our theoretical probabilities for steps 1 to 30
steps = np.arange(1, 31)
theoretical_probs = [f_theoretical(n) for n in steps]

#Simulate empirical probabilities for each step
def simulate_probability(n, trials=10000):
    hits = 0
    for _ in range(trials):
        pos = 1
        # Continue until we reach or pass step n.
        while pos < n:
            pos += np.random.choice([1, 2, 3])
        if pos == n:
            hits += 1
    return hits / trials

empirical_probs = [simulate_probability(n) for n in steps]

plt.figure(figsize=(8, 6))
plt.plot(steps, theoretical_probs, label='Theoretical  $\mathbb{P}$(landing on step $n$)',  linestyle='-')
plt.scatter(steps, empirical_probs, color='red', label='Empirical estimate', zorder=5)
plt.xlabel("Step Number")
plt.ylabel("Probability of landing exactly on step n")
plt.title("Probability of Landing Exactly on Step n")
plt.legend()
plt.grid(True)
plt.show()
```
